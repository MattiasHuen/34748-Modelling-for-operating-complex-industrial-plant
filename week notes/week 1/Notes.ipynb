## **Introduction to modelling for operating complex industrial plants**

A complex system is composed of many components which may interact with eachother. some of the features of the complex engineered systems are:
- Structural complexity - **many components**
- Procedural complexity - **many rules**
- Highly automated - **many hidden functions and interactions**
- Human operators - **are still in charge**
- Safety critical - **Don't make mistakes!**


**Definition of a system**
- Disctionary
    - System is a set of things working together as parts of amechanism or an interconnecting network; a complex whole.

- Eberhard Rechtin (system engineer)
    - A system is a construct or collection of different elements that together produce results not obtainable by the elements alone.
        - The elements or parts, can include people, hardware, software, facilities, policies, and documents; that is, all things required to produce systems-level results.
        - The results include system level qualities, characteristics, functions, behavior and performance.
        - The value added by the system as a whole, beyond that contributed independently by the parts, is primarily created by the relationship among the parts; that is, how they are interconnected.
